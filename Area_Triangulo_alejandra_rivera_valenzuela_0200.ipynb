# Formula para calcular Area
$A=(b*h)/2$
---
### datos de entrada
* b
* h
---

![Formula de Circurferencia](https://raw.githubusercontent.com/a25308051280200-max/Area-de-un-triangulo-/refs/heads/main/Area%20de%20triangulo%20diagrama%20.png)

```python
# Calculando el área de un triangulo
# Variables A, b, h,
# Fórmulas: A = (b * h)/2
# Capturar valor de a
import math
print("Calcular el área de un triángulo rectángulo")
b = float(input("Ingrese el valor de la base del triángulo: "))
h = float(input("Ingrese el valor de la altura del triángulo: "))

A = (b * h) / 2

print("El área del triángulo es: ", A)
print("Programa realizado por Alejandra Rivera Valenzuela NL = 45")
```
